```python
def open_file(filename):
    try:

        with open(filename, 'r') as file:

            contents = file.read()

            print("File contents:")

            print(contents)
    except FileNotFoundError:

        print("Error: File not found.")


file_name = input("Input a file name: ")


open_file(file_name)
```
