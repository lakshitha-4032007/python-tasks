```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 1000)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("Angle [radians]")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()


import matplotlib.pyplot as plt

height_ranges = [
    "140-145", "145-150", "151-156", "157-162", "163-168",
    "168-173", "173-178", "179-184", "185-190", "190-195"
]

people_counts = [2, 5, 15, 31, 46, 53, 45, 28, 21, 4]

plt.bar(height_ranges, people_counts)
plt.xlabel("Height Range (cm)")
plt.ylabel("Number of People")
plt.title("Number of People by Height Range")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


import matplotlib.pyplot as plt

X = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(X, y)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter Plot')
plt.show()


import matplotlib.pyplot as plt

labels = ['W', 'X', 'Y', 'Z']
sizes = [35, 25, 25, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.axis('equal')
plt.show()
```
