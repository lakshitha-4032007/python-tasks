```python
def find_discount(qty):
    if qty <= 10:
        return 0
    elif 11 <= qty <= 20:
        return 15
    else:
        return 20

def buy():
    ICode = int(input("Enter Item Code: "))
    Item = input("Enter Item Name: ")
    Price = float(input("Enter Price: "))
    Qty = int(input("Enter Quantity: "))

    Discount_percent = find_discount(Qty)
    total_price = Price * Qty
    Discount_amount = (Discount_percent / 100) * total_price
    Netprice = total_price - Discount_amount

    return ICode, Item, Price, Qty, Discount_amount, Netprice

def show_all(ICode, Item, Price, Qty, Discount, Netprice):
    print("\n--- BILL DETAILS ---")
    print("Item Code:", ICode)
    print("Item Name:", Item)
    print("Unit Price:", Price)
    print("Quantity:", Qty)
    print("Discount Amount:", round(Discount, 2))
    print("Net Price:", round(Netprice, 2))


ICode, Item, Price, Qty, Discount, Netprice = buy()
show_all(ICode, Item, Price, Qty, Discount, Netprice)

```
