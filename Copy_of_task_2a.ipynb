```python
import random

def guess_number():
    target_number = random.randint(1, 9)
    while True:
        try:
            user_guess = int(input("Guess a number between 1 and 9: "))
            if user_guess == target_number:
                print("Well guessed!")
                break
            else:
                print("Wrong guess. Try again.")
        except ValueError:
            print("Please enter a valid number.")

guess_number()
```
